# For customers who have more than 1 booking, which is the most booked route 
# (from_city to to_city)? 


```python
len(booking["customer_id"].unique())
```

```python
coust_count =booking['customer_id'].value_counts()
```

```python
coust_count
```

```python
custs_booking_multi = coust_count[coust_count > 1].index
```

```python
df_multiple_bookings = booking[booking['customer_id'].isin(custs_booking_multi)]
```

```python
df_multiple_bookings
```

```python
route_counts = df_multiple_bookings.groupby(['customer_id', 'from_city', 'to_city']).size().reset_index(name='count')
route_counts
```

```python
most_booked_routes = route_counts.loc[route_counts.groupby('customer_id')['count'].idxmax()]
```

```python
most_booked_routes
```

# Plot a heatmap displaying correlations of the numerical column and report which pair 
# of numerical columns in the bookings dataset, have the maximum correlation?

```python
numerical_df = booking.select_dtypes(include=['number'])
```

```python
corr=numerical_df.corr()
```

```python
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
max_corr_pair = (None, None)
max_corr_value = 0
for i in range(len(corr.columns)):
    for j in range(i+1, len(corr.columns)):
        if abs(corr.iloc[i, j]) > abs(max_corr_value):
            max_corr_value = corr.iloc[i, j]
            max_corr_pair = (corr.columns[i], corr.columns[j])
```

```python
max_corr_pair
```

# Which are the top 3 departure cities from where customers book mostly in advance, 
# provided that there have been at least 5 departures from that city? 

```python
city_counts = booking['from_city'].value_counts()
```

```python
city_counts
```

```python
cities_with_min_departures = city_counts[city_counts >= 5].index
```

```python
city_counts.info()
```

```python
filtered_df = booking[booking['from_city'].isin(cities_with_min_departures)]
```

```python
filtered_df
```

```python
avg_booking_times = filtered_df.groupby('from_city')['days_to_departure'].mean().reset_index()
```

```python
sort=avg_booking_times.sort_values(by='days_to_departure')
```

```python
print("top 3 departure cities")
print(sort.tail(3))
```
